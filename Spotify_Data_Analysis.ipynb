```python
# Importing needful Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Collecting the data

df_track = pd.read_csv("/content/drive/MyDrive/Datasets/Spotify/tracks.csv")
df_track
```

```python
# Droping the null values present in the name columns

df_track.dropna(inplace = True)
df_track.isnull().sum()
```

```python
df_track.head()
```

```python
# Getting the information from the dataset

df_track.info()
```

```python
# Fetching the top 10 least popular songs
sorted_df = df_track.sort_values('popularity', ascending = True).head(10)
sorted_df
```

These are the top 10 least popular songs in spotify.

```python
# Getting the statistical values for the dataset

df_track.describe().transpose()
```

```python
# Fetching the top 10 most popular songs which have popularity score over 90

most_pop = df_track.query('popularity>90').sort_values('popularity', ascending = False)
most_pop[:10]
```

```python
# Setting the index to as release date column

df_track.set_index('release_date', inplace = True)

df_track.index = pd.to_datetime(df_track.index)
df_track.head()
```

```python
df_track[['artists']].iloc[18]
```

```python
# Changing duration name and converting it into seconds
df_track['duration'] = df_track['duration_ms'].apply(lambda x: round(x/1000))
df_track.drop('duration_ms', inplace = True, axis = 1)
```

```python
df_track.head()
```

```python
# Checking the correlation of the variables

corr_df = df_track.drop(['key', 'mode', 'explicit'], axis = 1).corr(method = 'pearson')
plt.figure(figsize = (14,5))
heatmap = sns.heatmap(corr_df, annot = True, fmt = '.1g', vmin = -1, vmax = 1, center = 0, cmap = 'inferno', linewidths= 2)
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
# Taking a 4% sample out of main data to make regression plot using that sample data
sample_df = df_track.sample(int(0.004*len(df_track)))
print(len(sample_df))
```

```python
# Creating a regression plot for Loudness and Energy
plt.figure(figsize = (10,6))
sns.regplot(sample_df, y = 'loudness', x = 'energy', color = 'c').set(title = 'Loudness V/S Energy Correlation')
plt.show()
```

So we can clearly see that loundness and the energy are positively correlated to each other.

```python
# Creating a regression plot
plt.figure(figsize = (10,6))
sns.regplot(sample_df, y = 'popularity', x = 'acousticness', color = 'r').set(title = 'Popularity V/S Acousticness')
plt.show()
```

So we can see that the regression line is downward which tells if the popularity increases the acousticness decreases and acousticness increasess popularity decreases.

In simple terms they both are negatively correlated to each other.

```python
# Creating a new columns called year using release data column

df_track['dates'] = df_track.index.get_level_values('release_date')
df_track.dates = pd.to_datetime(df_track.dates)
years = df_track.dates.dt.year
```
